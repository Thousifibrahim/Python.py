```python
n = list(range(10, 100, 2))  # Creates a list of even numbers from 10 to 98.

for num in n:  # Loop through each number in the list.
    if num == 46 or num == 40:  # Check if the number is 46 or 40.
        print(f"You have reached {num}. Do you want to continue?")
        pro = input("Enter YES or NO: ").lower()  # Ask the user and convert input to lowercase.
        
        if pro == "yes":
            continue  # Skip to the next iteration of the loop.
        else:
            print("Stopping the process.", n)
            break  # Exit the loop.
else:
    print("All numbers have been processed.",n)  # This runs if the loop finishes without breaking.
```
