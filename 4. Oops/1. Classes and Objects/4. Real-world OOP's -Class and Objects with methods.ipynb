```python
import random

class Bank:
    bank_name="State Bank of India"
    branch="HMT branch"
    ifsc_code=789654123654
    
    def __init__(self,name,phone_no,mail_id):
        self.name=name
        self.phone_no=phone_no
        self.mail_id=mail_id
        self.deposit_money=0
        self.withdraw_money=0
        self.balance=0
        self.transfer=0
        self.passbook_=[]
        self.debit_card_number=0
        self.debit_card_number_password=0
        self.credit_card_number=0
        self.credit_card_number_password=0
        self.credit_card_limit_amount=100000
        
    def issue_debit_card(self):
        pin=int(input("Enter four digit pin"))
        random.seed(5)
        card = random.randint(1, 10000000000)
        self.debit_card_number=card
        if len(str(pin))<5 and len(str(pin))>3:
            self.debit_card_number_password=pin
        print(f"The {self.name} Debit Card number is {self.debit_card_number} and the password is {self.debit_card_number_password}")
        
    def issue_credit_card(self):
        pin=int(input("Enter four digit pin"))
        random.seed(5)
        card = random.randint(1, 10000000000)
        self.credit_card_number=card
        if len(str(pin))<5 and len(str(pin))>3:
            self.credit_card_number_password=pin
        print(f"The {self.name} Credit Card number is {self.credit_card_number} and the password is {self.credit_card_number_password} and the limit amount is {self.credit_card_limit_amount}")
        
    def credit_withdrawl(self,amount):
        pin=int(input("Enter the credit card pin"))
        if self.credit_card_number:
            if pin==self.credit_card_number_password:
                if self.credit_card_limit_amount>amount:
                    self.credit_card_limit_amount-=amount
                    print(f"The credit amount withdrawed is {amount} and balance credit card amount is {self.credit_card_limit_amount}")
                else:
                    print("Insufficient Credit Card Balance")
            else:
                print("Invalid Pin")
        else:
            print("Create Credit Card for further transaction")
            
    def deposit(self,amount):
        self.deposit_money=amount
        self.balance+=self.deposit_money
        print(f"The deposited amount is {self.deposit_money} and the balance is {self.balance}")
        self.passbook_.append(f"The amount deposited is {self.deposit_money} and the balance is {self.balance}")
        
    def withdraw(self,amount):
        pin=int(input("Enter the Debit Card Pin"))
        if self.debit_card_number:
            if pin==self.debit_card_number_password:
                if self.balance>amount:
                    self.withdraw_money=amount
                    self.balance-=self.withdraw_money
                    self.passbook_.append(f"The amount withdrawn is {self.withdraw_money} and the balance is {self.balance}")
                    print(f"The withdrawed amount is {amount} and the remaining balance is {self.balance}")
                else:
                    print("Insufficient Balance")
            else:
                print("Invalid Pin")
        else:
            print("Invalid Debit card")
        
    def check_balance(self):
        print("The Balance amount is",self.balance)
        
    def passbook(self):
        print("Transaction History")
        for entry in self.passbook_:
            print(entry)
    
    def __str__(self):
        return f"Account holder: {self.name}, Phone: {self.phone_no}, Balance: {self.balance}"
    
jenny=Bank("Jensilin",6363784224,"jensilinjino@gmail.com")
jenny.deposit(1000000)
jenny.deposit(1000)
jenny.issue_debit_card()
jenny.issue_credit_card()
jenny.withdraw(500)
jenny.credit_withdrawl(10000)
```

# UPI

Class UPI:

        Class Members:
                Application name - Phonepe


        Object Members:
                Upi object should be the object in Bank class and should have debit card
                Creating upi id 
                Generating Upi pin (6 digits)

```python
class UPI:
    application_name="PhonePe"
    
    def __init__(self,bank_account,upi_id,upi_pin):
        
        if bank_account is None or not isinstance(bank_account,Bank):
            print("Please create bank account first")
            return
        
        if bank_account.debit_card_number is None:
            print("Upi Id cannot be generated without debit card")
            return
        self.bank_account=bank_account
        self.upi_id=upi_id
        self.upi_pin=upi_pin
        print(f"The generated upi id is {upi_id} and upi pin is {upi_pin}")
        
    def make_payment(self,amount):
        pin=int(input("Enter the UPI Pin: "))
        if pin==self.upi_pin:
            if self.bank_account.balance>=amount:
                self.bank_account.balance-=amount
                print(f"Payment of {amount} successful! Remaining amount is {self.bank_account.balance}")
            else:
                print("Insufficient Balance")
        else:
            print("Incorrect UPI PIN")
    
jenny_upi=UPI(jenny,upi_id="jensilin",upi_pin=2003)
jenny_upi.make_payment(500)
```
