```python
#WAP to check whether the first value of a list collection belongs to which collection datatype

a = eval(input("Enter the list"))
if type(a[0])==list:
    print("List")
elif type(a[0])==dict:
    print("Dict")
elif type(a[0])==set:
    print("Set")
elif type(a[0])==tuple:
    print("Tuple")    
else:
    print("It is not a collection data type")
```
