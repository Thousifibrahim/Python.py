```python
# WAP TO PRINT THE LAST VALUE OF LIST COLLECTION ONLY IF IT IS A STRING DATATYPE ,HAVING MIDDLE VALUE AS AN UPPERCASE

a=eval(input("Enter the list value :" ))

b = a[-1]
if type(b) == str :
    if len(b)%2!=0:
        d = len(b) // 2
        if 'A'<= b[d]<='Z':
            print(b)
        else :
            print("Middle value is in lower case")
    else :
            print("No middle value")
else :
    print("The last value is not string ")
```

```python

#--------------------#Assignment-------------------------------------------
#  consider marks of a student in six subjects , Find the percentage of the marks only
# if he or she has scored more than thirty five 
# # 35 in all the subjects 
# # print the results as 

sub1 = int(input("Enter marks for subject 1: "))
sub2 = int(input("Enter marks for subject 2: "))
sub3 = int(input("Enter marks for subject 3: "))
sub4 = int(input("Enter marks for subject 4: "))
sub5 = int(input("Enter marks for subject 5: "))
sub6 = int(input("Enter marks for subject 6: "))

if sub1 > 35 and sub2 > 35 and sub3 > 35 and sub4 > 35 and sub5 > 35 and sub6 > 35:
    total = sub1 + sub2 + sub3 + sub4 + sub5 + sub6  
    per = (total / 600) * 100  
    print("Percentage:", per)

    # Grading system
    if 35 >= per <= 45:
        print("Grade: D")
    elif 46 >= per <= 55:
        print("Grade: C")
    elif 56 >= per <= 65:
        print("Grade: B")
    elif 66 >= per <= 75:
        print("Grade: First Class")
    elif 76 >= per <= 85:
        print("Grade: First Class Exemplary")
    elif 86 >= per <= 100:
        print("Grade: Distinction")
else:
    print("The student scored less than 35 in one or more subjects.")
    
```
