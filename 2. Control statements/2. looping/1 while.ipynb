```python
# WAP program to check wheather the ineteger number is AMR STRONG 
# 1234 , len is 4 so cube becomes 4 , for each  number 1 power 4, 2 power 3.

n  =  int(input("Enter the number "))
ns = str(n)                                                           # for tracing in in index
cal = 0                                                               # seperate ouput
i = 0 
while i < len(ns) :
     cal =cal + int(ns[i])**len(ns)                                   #cal updated power  + ns number + len
     i+=1
print(cal)

if cal == n:
    print("The number is an Armstrong number.")
else:
    print("The number is not an Armstrong number.")
```
