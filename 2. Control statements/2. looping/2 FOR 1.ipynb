```python
#---------------------------------------WEEKEND----------------
#  shed ge ba 
# Output  : ShEd Ge Ba  --------------------- EVEN INDEX CAPITAl

st = input("Enter the string :")
x =st.split()

out = ' ' 

for i in x :
    ind = ' '
    for j in range(len(i)):
        if j%2==0 and i[j].islower():
            ind=ind+i[j].upper()
        else:
            ind+=i[j]
            
    out+=ind+ ' '
print(out)
```

```python
# WAP to store letter and its count of occurence in the form of KeY value pairs of a string collection ----------------------------------------------- IMP -----------


st = 'aaabbaavvcca'

out = ' '
for i in st :
    if i not in out :
        c = 0
        for j in st :
            if  i==j:
                c+=1
        out = out+i +str(c)
print(out)
```
