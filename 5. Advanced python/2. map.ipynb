```python
length = lambda s : len(s)

st=input("Enter the string to count to the word present:")

spl=st.split()

cont = map(length,spl)

print(list(cont))
 
```

```python
# WAP to store the factorials of  all the numbers between the range 1 to 5 

def fact(n):
    f=1
    for i in range(1,n+1):
        f*=i
    return f

ser_fact=map(fact,range(1,6))
print(list(ser_fact))
```
