    LAMDA
        It is a keyword which acts like a anonymous function in python to perform simple operations.
        var= lamda args:expression
        print(var(values))


```python
# WAP which returns or prints the square of a number if it is odd number , returns cube of a number 
# if it is even 

Syntax : var = lambda value/args 

sq = lambda r : r**2 if r%2==1 else r**3
print(sq(3))
```

```python
# WAP to find greatest of two integer numbers

gre = lambda a,b : print(a) if a>b else print(b)
print(gre(8,9)) 
```

```python
#WAP to reverse the value, if not possible to reverse print the value as it is
rev=lambda n:n[::-1] if type(n) in[list,str,tuple] else n
print(rev(1254))
```
