# Decorator

Decorator 
        It is a function which is used to  add additional features to the existing function without modifying it.
        The decortor is used on the functions which have common pre and post task.
        
        It is of two types:
            Inbuilt decorator
            User-defined decorator

        Inbuilt decorator : 
                These are pre-defined decorators to perform some task
                example: @classmethod @staticmethod @abstractmethod

        User-defined decorators
                These are the decorators which are created based on user requirements.
                Syntax: 
                    def decor_name(fname):
                        def inner(*args,**args):
                            pre task
                            fname(*args,**kwargs):
                                post task
                            
                        return inner
                    
                    @decorator name
                    def fname(args):
                        S.B


        NOTE: A DECORATOR CAN HAVE ONLY PRE-TASK OR POST-TASK OR INCLUDING BOTH PRE-TASK AND POST-TASK

```python
#---------------IMPO---------------------

# WAP to calculate the TOTAl execution time of any program 

#-----------------USING TIME MODULE-------------------------

import time

def timee(func):
    def inner(*args, **kwargs):
        start = time.time()
        func(*args, **kwargs)
        end = time.time()
        print(f"Execution time: {end - start} seconds")
        return result
    return inner

@timee
def mul():
    for i in range(1, 500):
        yield i
mul()
```

```python
#--------------------IMPO-------------------------

# WAP to create a decorator which gives 5 second deplay defore executing the program
# NOTE - not mandatory to have PRE-and-POST

#---USING IN_BUILT TIME DEPLAY FUNCTION OF time.sleep(deplay sec)----------------------

import time
def delay(func):
    def inner(*args, **kwargs):
        time.sleep(5)
        func(*args, **kwargs)
    return inner

@delay
def series():
    for i in range(10, 100):
        print(i)
series()
```

```python
# ----------------------------------IMPO-------------------------\\
    
    
# Create a  decorator which returns positive values always when an airthmethic operation 
# is performed

#  ABSOLUTE _VALUE Return Function


def absolute(func):
    def inner(*args, **kwargs):
        res =func(*args, **kwargs)
        print(abs(res))
    return inner

@absolute
def add(a, b):
    return a + b

add(-2,4)

@absolute
def subtract(a, b,c):
    return a - b  -c
subtract(3,+3,8)
```

```python
def instagram(func):
    def inner(*args,**kwargs):
        print("Logged In")
        func(*args,**kwargs)
        print("logged out")
    return inner

@instagram
def Adhi():
    print("Seee reels")
    print("Seee story")
    print("Seee others")
    print("Seee see see")

Adhi()                  #using decorator we get logged in and logged out message
```

```python
#WAP which calculates the total execution time of a program


def record(rec):
    def inner(*args,**kwargs):
        import time
        start=time.time()
        rec(*args,**kwargs)
        end=time.time()
        print(end-start)
    return inner

@record
def rec():
    for i in range(0,100):
        print(i)
        
rec()
```

```python
#Create a decorator which return positive values when an arthmethic operation is performed

def return_positive(arith):
    def inner(*args,**kwargs):
        res=arith(*args,**kwargs)
        print(res)
    return inner

@return_positive
def add(a,b):
    print(a+b)

add(-10,5)
```
